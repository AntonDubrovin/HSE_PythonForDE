```python
folder = 'customer_database/src/'
```

```python
import pandas as pd
```

```python
customer = pd.read_csv(folder + 'customer.csv', sep=';')
customer.head()
```

```python
contact_preference = pd.read_csv(folder + 'contact_preference.csv', sep=';')
contact_preference.head()
```

```python
customer_preferenced_contact_type = pd.read_csv(folder + 'customer_preferenced_contact_type.csv', sep=';')
customer_preferenced_contact_type.head()
```

```python
customers_segmetns = pd.read_csv(folder + 'customers_segmetns.csv', sep=';')
customers_segmetns.head()
```

```python
location = pd.read_csv(folder + 'location.csv', sep=';')
location.head()
```

```python
product_feedback = pd.read_csv(folder + 'product_feedback.csv', sep=';')
product_feedback.head()
```

```python
product = pd.read_csv(folder + 'product.csv', sep=';')
product.head()
```

```python
purchase = pd.read_csv(folder + 'purchase.csv', sep=';')
purchase.head()
```

```python
segment = pd.read_csv(folder + 'segment.csv', sep=';')
segment.head()
```

```python
service_feedback = pd.read_csv(folder + 'service_feedback.csv', sep=';')
service_feedback.head()
```

```python
all_data = pd.merge(customer, customer_preferenced_contact_type, left_on='id', right_on='customer_id', left_index=False, right_index=False)
all_data = pd.merge(all_data, contact_preference, left_on='contact_type_id', right_on='contact_id', left_index=False, right_index=False)
all_data = pd.merge(all_data, customers_segmetns, left_on='customer_id', right_on='customer_id', left_index=False, right_index=False)
all_data = pd.merge(all_data, segment, left_on='segment_id', right_on='id', left_index=False, right_index=False)
all_data = pd.merge(all_data, location, left_on='location_id', right_on='location_id', left_index=False, right_index=False)
all_data = pd.merge(all_data, purchase, left_on='customer_id', right_on='user_id', left_index=False, right_index=False, suffixes=('_left', '_purchase'))
all_data = pd.merge(all_data, product, left_on='product_id', right_on='id', left_index=False, right_index=False, suffixes=('_left', '_right'))
all_data = pd.merge(all_data, product_feedback, left_on='id_purchase', right_on='id', left_index=False, right_index=False, suffixes=('_left1', '_right1'))
all_data.head()
```
